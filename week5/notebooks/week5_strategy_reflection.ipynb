## 1. How did hierarchical feature learning influence my strategy?

Early weeks were about figuring out which parts of each function even had any signal. That is the coarse level -- just finding regions that are not zero or not completely bad. Later weeks are about refining within those regions. That matches how a neural network works -- early layers find broad patterns, later layers get more specific.

The clearest example is F5. The first two weeks told me that high x3 and x4 values matter. Week 4 showed that pulling x1 lower at the same time pushed the output from 1192 to 3531. That combination was only findable once there were enough observations for the NN to learn it. The GP missed it completely because it uses the same smoothness assumption everywhere.

For Week 5 I am using this same idea to decide which functions need more exploration and which need tighter refinement. F5, F7 and F8 are at the fine layer now -- small moves near the best known point. F1 and F4 are still at the coarse layer -- we do not yet know where the good region is, so the GP explores more broadly.

---
## 2. What parallels do I see with AlexNet and ImageNet breakthroughs?

F5 jumping from 1192 to 3531 in Week 4 is the closest thing I have had to that kind of leap. For two rounds in a row the output was stuck at 1192. Nothing pointed to a jump being possible. The NN found a different combination of inputs that the GP had not considered and it worked.

But Week 4 also had the opposite. F4 went to -48, which is the worst result in the whole project. That happened because the NN gradient ascent kept pushing toward the corners of the input space even though the corners are clearly bad. So the same tool that caused the breakthrough in F5 caused a disaster in F4.

The AlexNet lesson that applies here is that progress is not uniform. Some functions respond well to the new approach and some do not. Trying to use the same method on all eight functions when they are behaving so differently is the mistake. This week I split them: NN where it has helped, GP only where the NN has been sending queries to bad places.

---
## 3. Did I face exploration vs exploitation trade-offs?

Yes, and it was the main decision this week.

For F5, F3 and F8 the results have been getting better. The right move is to stay close to what is already working and make small adjustments. That is exploitation. The trust region is tight -- F5 uses ±0.05 around the Week 4 best point.

For F1, F2, F4 and F6 the NN was sending queries to bad regions. F4 went to -48 when the gradient ascent pushed to the boundary corners [1, 1, 1, 0]. That is the deepest the network has gone wrong. Switching to GP with EI means the model only recommends a point if it genuinely expects to beat the current best. It is more careful.

The hardest trade-off was F7. Week 3 gave 1.235, which is the best result for that function. Week 4 gave 0.998, which is worse. So this week I go back to the Week 3 best as the starting point rather than the Week 4 one. Sometimes the most recent result is not the best guide for where to go next.

## 4. Which neural network building block helped me think differently?

Gradients were the most useful concept this week, but for a different reason than before.

In Week 4, all eight functions used the NN with gradient ascent. Looking at the results, the gradient ascent was clearly working well for F5, F7 and F8 -- these are the functions where it found good regions the GP had missed. For F1, F2, F4 and F6 the gradient ascent went to boundaries and corners and made things worse.

This told me that the gradient signal from the NN is only trustworthy when the NN has learned something reliable. With only 14 observations in F1 and 34 in F4, the NN can be confidently wrong. The GP is more honest -- it tells you when it does not know something through its uncertainty estimate. The NN has no uncertainty, so it will follow a gradient into a bad region without any warning.

The lesson: gradients are powerful but you need enough data for them to point in a useful direction. For small datasets the GP uncertainty is more valuable than the NN gradient.

## 5. Is my approach closer to rapid prototyping or production-ready design?

Rapid prototyping. The strategy is different for every function and changes every week based on what the results say.

This week PyTorch is only used for F3, F5, F7 and F8 -- the functions where the NN is actually helping. For F1, F2, F4 and F6 the notebooks do not import PyTorch at all. Using a framework just because it was used last week would be the wrong choice when the evidence says it is not working for those functions.

The SVM also changed this week. Before, it printed a warning if the proposed query looked poor but never changed the query. That is not useful -- it is just extra output. Now the SVM scores both the NN gradient candidate and the GP candidate and picks whichever one it thinks is more likely to be in a good region. If the primary candidate is flagged as poor, the other one is tried instead. The SVM is now actually doing something.

A production-ready system would have the same pipeline for all functions. What I have is the opposite -- each function has its own setup based on what it has shown so far.

## 6. How did real-world applications change how I measure success?

Week 4 was the worst week by the simple metric of did the numbers go up. F4 went to -48, F6 went to -0.98, F1 went negative, F2 went negative. Four functions got worse.

But Week 4 was very useful. It showed clearly which functions the NN cannot be trusted on yet. That is information. Without those bad results I would have kept using the NN on all eight functions and probably made the same mistakes in Week 5.

The sports analytics point from the module is that in real applications a bad outcome is not a failed experiment -- it is data. A team that loses and learns why they lost is better positioned than a team that wins without understanding why.

For this project I now measure each week on two things: did any outputs improve, and did I learn something that changes the strategy. Week 4 scored zero on the first and high on the second. That is why Week 5 has a different setup for half the functions. The bad results in Week 4 made Week 5 more targeted.
